```python
import pydicom
import matplotlib.pyplot as plt
import glob
import numpy as np
```

```python
raw_path = '/Users/mona/Library/CloudStorage/Box-Box/Animals/Paris/Paris_wk8+2/DCE/T1   4   6   8  10  12  14  16  26  28  30  32min/Images'
MOCO_path = '/Users/mona/Library/CloudStorage/Box-Box/Animals/Paris/Paris_wk8+2/DCE_MOCO/T1   4   6   8  10  12  14  16  26  28  30  32min/Images'
```

```python
def plot_raw_moco_diff(raw_path, MOCO_path, label, VMAX = 100):
    cxm_rsq_raw = glob.glob(f"{raw_path}/{label}/*.ima")[0]
    cxm_rsq_moco = glob.glob(f"{MOCO_path}/{label}/*.ima")[0]

    cxm_rsq_raw_img = pydicom.dcmread(cxm_rsq_raw).pixel_array.astype(np.float32)
    cxm_rsq_moco_img = pydicom.dcmread(cxm_rsq_moco).pixel_array.astype(np.float32)

    
    fig = plt.figure(figsize=(12, 3))
    fig.suptitle(label)
    fig.add_subplot(1, 3, 1)
    ax1 = plt.gca()
    im1 = ax1.imshow(cxm_rsq_raw_img, cmap='jet', vmin=0, vmax=VMAX)
    ax1.axis('off')
    ax1.set_title('Raw')
    plt.colorbar(im1, ax=ax1)

    fig.add_subplot(1, 3, 2)
    ax2 = plt.gca()
    im2 = ax2.imshow(cxm_rsq_moco_img, cmap='jet', vmin=0, vmax=VMAX)
    ax2.axis('off')
    ax2.set_title('MOCO')
    plt.colorbar(im2, ax=ax2)

    fig.add_subplot(1, 3, 3)
    ax3 = plt.gca()
    im3 = ax3.imshow(cxm_rsq_moco_img - cxm_rsq_raw_img, cmap='jet', vmin=-VMAX/2, vmax=VMAX/2)
    ax3.axis('off')
    ax3.set_title('Diff')
    plt.colorbar(im3, ax=ax3)
    plt.tight_layout()
    return fig
```

```python
fig = plot_raw_moco_diff(raw_path, MOCO_path, label='cxm_PS_T1   4   6   8  10  12  14  16  26  28  30  32min', VMAX = 2000)
fig = plot_raw_moco_diff(raw_path, MOCO_path, label='cxm_Ve_T1   4   6   8  10  12  14  16  26  28  30  32min', VMAX = 8000)
fig = plot_raw_moco_diff(raw_path, MOCO_path, label='cxm_Vp_T1   4   6   8  10  12  14  16  26  28  30  32min', VMAX = 2000)
fig = plot_raw_moco_diff(raw_path, MOCO_path, label='cxm_Rsq_T1   4   6   8  10  12  14  16  26  28  30  32min', VMAX = 100)
```
