## Check Registration

+ Check the Signal intensity in different region among slices before and after the registration


```python
import numpy as np
import matplotlib.pyplot as plt
folder = '/Users/mona/Library/CloudStorage/Box-Box/Animals/Paris/Paris_D8/registration'
stage1 = np.load(f"{folder}/register_1.npy")
stage2 = np.load(f"{folder}/register_2.npy")
```

```python
def plot_slices(imgs, height=5, colorbar=(0, 2000)):
    """ Function to plot row of image slices """
    n_slices = imgs.shape[-1]
    fig, axes = plt.subplots(1, n_slices, figsize=(n_slices*height, height))
    for n, ax in zip(range(0, n_slices), axes):
        ax.imshow(imgs[..., n], cmap="gray", origin="upper", vmin=colorbar[0], vmax=colorbar[1])
        ax.axis("off")
        

    fig.tight_layout()
    plt.show()
```

```python
plot_slices(stage1)
plot_slices(stage2)
```

```python
# compare the signal intensity
def plot_SI(img1, img2, position):
    x, y = position
    data1 = img1[x, y, :]
    data2 = img2[x, y, :]
    plt.plot(data1, label='Original')
    plt.plot(data2, label='After Registration')
    plt.title(f'Pixel SI among timesteps at ({x}, {y})')
    plt.legend()
```

```python
n_rows, n_cols, n_slices = stage1.shape
plot_SI(stage1, stage2, position=(n_rows//2, n_cols//2))
```
